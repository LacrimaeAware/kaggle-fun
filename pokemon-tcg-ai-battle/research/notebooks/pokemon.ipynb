```python
import sys
sys.path.append("/kaggle/input/competitions/pokemon-tcg-ai-battle/Sample Submission")
from cg.api import all_card_data, all_attack
import pandas as pd

all_cards = all_card_data()
all_attacks = all_attack()
attack_map = {a.attackId: a for a in all_attacks}
df = pd.read_csv("/kaggle/input/competitions/pokemon-tcg-ai-battle/EN_Card_Data.csv")

# 1. BASIC покемоны с HP > 100
print("=" * 60)
print("BASIC POKEMON HP > 100:")
print("=" * 60)
basics = []
for c in all_cards:
    if c.cardType == 0 and c.basic and c.hp > 100:
        row = df[df['Card ID'] == c.cardId]
        w = ''
        r = ''
        rc = c.retreatCost
        if c.weakness > 0:
            w = str(c.weakness)
        if c.resistance > 0:
            r = str(c.resistance)
        basics.append((c.cardId, c.name, c.hp, c.energyType, rc, c.ex))
basics.sort(key=lambda x: x[2], reverse=True)
for cid, name, hp, etype, rc, ex in basics[:25]:
    ex_str = " EX" if ex else ""
    print(f"  [{cid}] {name}{ex_str}: HP={hp} EType={etype} Retreat={rc}")

# 2. SUPPORTER
print("\n" + "=" * 60)
print("SUPPORTER CARDS:")
print("=" * 60)
for c in all_cards:
    if c.cardType == 3:
        row = df[df['Card ID'] == c.cardId]
        eff = str(row.iloc[0].get('Effect Explanation', ''))[:150] if len(row) > 0 else ''
        print(f"  [{c.cardId}] {c.name}: {eff}")

# 3. ITEM (первые 25)
print("\n" + "=" * 60)
print("ITEM CARDS:")
print("=" * 60)
cnt = 0
for c in all_cards:
    if c.cardType == 1:
        row = df[df['Card ID'] == c.cardId]
        eff = str(row.iloc[0].get('Effect Explanation', ''))[:120] if len(row) > 0 else ''
        ace = " ACE" if c.aceSpec else ""
        print(f"  [{c.cardId}] {c.name}{ace}: {eff}")
        cnt += 1
        if cnt >= 30:
            break

# 4. POKEMON TOOL
print("\n" + "=" * 60)
print("POKEMON TOOLS:")
print("=" * 60)
for c in all_cards:
    if c.cardType == 2:
        row = df[df['Card ID'] == c.cardId]
        eff = str(row.iloc[0].get('Effect Explanation', ''))[:120] if len(row) > 0 else ''
        ace = " ACE" if c.aceSpec else ""
        print(f"  [{c.cardId}] {c.name}{ace}: {eff}")

# 5. Атаки нашей деки
print("\n" + "=" * 60)
print("ATTACKS IN SAMPLE DECK:")
print("=" * 60)
for cid in [721, 722, 723]:
    c = next((x for x in all_cards if x.cardId == cid), None)
    if c:
        print(f"\n{c.name} (ID={cid}, HP={c.hp}):")
        for aid in c.attacks:
            a = attack_map.get(aid)
            if a:
                print(f"  [{aid}] {a.name}: dmg={a.damage}, energies={a.energies}")
```

```python
import sys, random
sys.path.append("/kaggle/input/competitions/pokemon-tcg-ai-battle/Sample Submission")
from cg.game import battle_start, battle_select, battle_finish
from cg.api import to_observation_class, OptionType, SelectContext, all_card_data, all_attack

all_cards = all_card_data()
card_info = {c.cardId: c for c in all_cards}
all_attacks = all_attack()
attack_info = {a.attackId: a for a in all_attacks}

new_deck = [
    722, 722, 722, 722,
    723, 723, 723, 723,
    721, 721,
    1064, 1064,
    1145, 1145, 1145, 1145,
    1086, 1086, 1086, 1086,
    1097, 1097,
    1079, 1079,
    1205, 1205, 1205, 1205,
    1227, 1227, 1227, 1227,
    1235, 1235, 1235, 1235,
    1182, 1182,
    1212, 1212,
    1158, 1157, 1157, 1175,
    3, 3, 3, 3, 3, 3, 3, 3, 3, 3, 3, 3, 3, 3, 3, 3,
]

SEARCH_CARDS = {1145, 1205}
DRAW_CARDS = {1227}
ENERGY_ACCEL = {1235}
BENCH_FILL = {1086}
TOOLS = {1158, 1157, 1175}
BASIC_POKEMON = {722, 721, 1064}

def agent(obs_dict):
    obs = to_observation_class(obs_dict)
    if obs.select is None:
        return new_deck
    
    options = obs.select.option
    mc = obs.select.maxCount
    ctx = obs.select.context
    player_idx = obs.current.yourIndex
    hand = obs.current.players[player_idx].hand
    
    attacks, evolves, attaches, plays, ends = [], [], [], [], []
    cards, yes, no, numbers = [], [], [], []
    
    for i, opt in enumerate(options):
        t = int(opt.type)
        if t == OptionType.ATTACK: attacks.append(i)
        elif t == OptionType.EVOLVE: evolves.append(i)
        elif t == OptionType.ATTACH: attaches.append(i)
        elif t == OptionType.PLAY: plays.append(i)
        elif t == OptionType.END: ends.append(i)
        elif t == OptionType.CARD: cards.append(i)
        elif t == OptionType.YES: yes.append(i)
        elif t == OptionType.NO: no.append(i)
        elif t == OptionType.NUMBER: numbers.append(i)
    
    def safe_return(indices):
        """Гарантирует что индексы валидны"""
        result = []
        for idx in indices:
            if 0 <= idx < len(options):
                result.append(idx)
        if len(result) < mc:
            # Добавляем недостающие из доступных
            for i in range(len(options)):
                if i not in result:
                    result.append(i)
                    if len(result) >= mc:
                        break
        return result[:mc] if mc > 1 else [result[0]]
    
    def get_hand_card(idx):
        if hand and idx < len(hand):
            return hand[idx].id
        return None
    
    def attack_damage(opt_idx):
        opt = options[opt_idx]
        aid = opt.attackId
        atk = attack_info.get(aid)
        return atk.damage if atk else 0
    
    # SETUP
    if ctx == SelectContext.SETUP_ACTIVE_POKEMON:
        for i in cards:
            cid = get_hand_card(options[i].index)
            if cid == 722:
                return safe_return([i])
        for i in cards:
            cid = get_hand_card(options[i].index)
            if cid in BASIC_POKEMON:
                return safe_return([i])
        if cards:
            return safe_return([cards[0]])
    
    # TO_HAND
    if ctx == SelectContext.TO_HAND:
        for i in cards:
            cid = get_hand_card(options[i].index)
            if cid == 723:
                return safe_return([i])
        if cards:
            return safe_return([cards[0]])
    
    # MAIN
    if evolves:
        return safe_return([evolves[0]])
    
    if plays:
        for priority_set in [SEARCH_CARDS, BENCH_FILL, DRAW_CARDS, ENERGY_ACCEL, TOOLS]:
            for i in plays:
                cid = get_hand_card(options[i].index)
                if cid in priority_set:
                    return safe_return([i])
        return safe_return([plays[0]])
    
    if attacks:
        for i in attacks:
            if attack_damage(i) >= 100:
                return safe_return([i])
        for i in attacks:
            if attack_damage(i) > 0:
                return safe_return([i])
    
    if attaches:
        return safe_return([attaches[0]])
    
    if ends:
        return safe_return([ends[0]])
    
    if cards: return safe_return([cards[0]])
    if yes: return safe_return([yes[0]])
    if no: return safe_return([no[0]])
    if numbers: return safe_return([numbers[0]])
    
    return safe_return([0])

# Тест с защитой
wins = 0
errors = 0
for game in range(10):
    try:
        obs, start = battle_start(new_deck, new_deck)
        while obs['current']['result'] < 0:
            if obs['current']['yourIndex'] == 0:
                choice = agent(obs)
                if not choice or choice[0] >= len(obs['select']['option']):
                    print(f"BAD CHOICE: {choice}, max={len(obs['select']['option'])}")
                    errors += 1
                    break
            else:
                obs_c = to_observation_class(obs)
                choice = random.sample(range(len(obs_c.select.option)), obs_c.select.maxCount)
            obs = battle_select(choice)
        if obs['current']['result'] == 0:
            wins += 1
        battle_finish()
    except Exception as e:
        errors += 1
        print(f"Error in game {game}: {e}")

print(f"Wins: {wins}/10, Errors: {errors}")
```
