# UMUD EXP59 sleep matrix: segmentation-first overnight run

Import this notebook when you want Kaggle to work while you sleep.

Run setup:
1. Add the competition input: **UMUD Challenge: Muscle Architecture in Ultrasound Data**.
2. Set **Accelerator = GPU** and **Internet = On**.
3. Run All.

This notebook is deliberately focused:
- **Train what we actually have labels for:** the two supervised mask tasks.
- **Do not pretend text/ruler/field are supervised labels:** those remain deterministic diagnostics in this run.
- **Keep scale visible:** every run saves its `calibration_measurement_debug.csv` with a run-specific name.
- **Produce several submit-ready CSVs:** each run copies its submission to a unique filename, then bundles everything into one zip.

Planned runs, in order:
1. `seg59_02_highres_512_unet` - conservative 512 px U-Net.
2. `seg59_03_highres_512_unetplusplus` - stronger decoder/skip fusion, safer batch size.
3. `seg59_06_highres_512_unet_strong_aug` - same high-res U-Net, stronger augmentation to test robustness.

Expected runtime: roughly 2-5 hours total depending on Kaggle GPU/package lag. If you only want one or two runs, change `MAX_RUNS` in the run cell.


```python
# Fast data preflight. This fails before package install/training if the competition input is not attached.
from pathlib import Path
import os

IMG_EXTS = ('.tif', '.tiff', '.png', '.jpg', '.jpeg', '.bmp')
LEAVES = {
    'apo_img': ['apo_images_new_model_v1'],
    'apo_msk': ['apo_masks_new_model_v1'],
    'fasc_img': ['fasc_images_new_model_v1'],
    'fasc_msk': ['fasc_masks_new_model_v1'],
    'test': ['test_set_v2', 'test_images_v2'],
}
TERMINAL_LEAVES = {'apo_images_new_model_v1', 'apo_masks_new_model_v1', 'fasc_images_new_model_v1', 'fasc_masks_new_model_v1', 'test_set_v2'}

def list_images(d):
    try:
        return [p for p in d.iterdir() if p.is_file() and p.suffix.lower() in IMG_EXTS]
    except OSError:
        return []

def print_tree(root, max_depth=3, depth=0, budget=None):
    budget = [300] if budget is None else budget
    try:
        entries = sorted(root.iterdir())
    except OSError:
        return
    for p in entries:
        if budget[0] <= 0:
            return
        budget[0] -= 1
        print('  ' * depth + ('[D] ' if p.is_dir() else '    ') + p.name)
        if p.is_dir() and depth < max_depth:
            print_tree(p, max_depth, depth + 1, budget)

def index_root(root):
    index = {}
    for dirpath, dirnames, _files in os.walk(root):
        for dn in dirnames:
            index.setdefault(dn, []).append(Path(dirpath) / dn)
        dirnames[:] = [d for d in dirnames if d not in TERMINAL_LEAVES]
    return index

def resolve_key(index, key):
    for leaf in LEAVES[key]:
        cands = index.get(leaf, [])
        if key == 'test':
            cands = sorted(cands, key=lambda d: len(list_images(d)), reverse=True)
            cands = [d for d in cands if list_images(d)]
        if cands:
            return cands[0]
    return None

roots = sorted(p for p in Path('/kaggle/input').iterdir() if p.is_dir()) if Path('/kaggle/input').exists() else []
print('Mounted roots:')
for root in roots:
    print(' -', root)

resolved = None
best_n = -1
for root in roots:
    idx = index_root(root)
    dirs = {key: resolve_key(idx, key) for key in LEAVES}
    n = sum(v is not None for v in dirs.values())
    if n > best_n:
        resolved = dirs
        best_n = n
resolved = resolved or {key: None for key in LEAVES}

print('\nResolved data folders:')
for key, path in resolved.items():
    count = len(list_images(path)) if path else 0
    print(f'  {key:8s} -> {path} ({count} images)' if path else f'  {key:8s} -> None')
missing = [key for key, path in resolved.items() if path is None]
if missing:
    print('\nMounted tree for debugging:')
    for root in roots or [Path('/kaggle/input')]:
        print(f'[root] {root}')
        print_tree(root)
    raise RuntimeError(f'Missing UMUD folders: {missing}. Attach the competition input, then Run All again.')

assert len(list_images(resolved['test'])) == 309, f"expected 309 test images, found {len(list_images(resolved['test']))}"
print('\nPreflight OK: competition data is mounted and test set has 309 images.')
```

```python
# Run the deliberate sleep matrix. Set MAX_RUNS to 1/2/3 if you want to shorten it.
import json
import os
import shutil
import subprocess
import sys
import time
from pathlib import Path

import pandas as pd

WORK = Path('/kaggle/working')
RUNS = [
    {
        'run_id': 'seg59_02_highres_512_unet',
        'reason': 'conservative high-resolution baseline; best first sanity candidate',
        'env': {
            'UMUD_EPOCHS': '18',
            'UMUD_IMG_SIZE': '512',
            'UMUD_TTA_EXTRA_SIZE': '640',
            'UMUD_MODEL_ARCH': 'unet',
            'UMUD_MODEL_ENCODER': 'resnet34',
            'UMUD_MODEL_ENCODER_WEIGHTS': 'imagenet',
            'UMUD_LOSS_MODE': 'dice_bce',
            'UMUD_AUG_LEVEL': 'light',
            'UMUD_BATCH_SIZE': '6',
            'UMUD_CLAHE': '0',
            'UMUD_FASC_POS_WEIGHT': '0',
            'UMUD_WEIGHTS_TAG': 'seg59_02',
        },
    },
    {
        'run_id': 'seg59_03_highres_512_unetplusplus',
        'reason': 'stronger decoder/skip fusion for finer sparse mask detail; lower batch for memory safety',
        'env': {
            'UMUD_EPOCHS': '18',
            'UMUD_IMG_SIZE': '512',
            'UMUD_TTA_EXTRA_SIZE': '640',
            'UMUD_MODEL_ARCH': 'unetplusplus',
            'UMUD_MODEL_ENCODER': 'resnet34',
            'UMUD_MODEL_ENCODER_WEIGHTS': 'imagenet',
            'UMUD_LOSS_MODE': 'dice_bce',
            'UMUD_AUG_LEVEL': 'light',
            'UMUD_BATCH_SIZE': '4',
            'UMUD_CLAHE': '0',
            'UMUD_FASC_POS_WEIGHT': '0',
            'UMUD_WEIGHTS_TAG': 'seg59_03',
        },
    },
    {
        'run_id': 'seg59_06_highres_512_unet_strong_aug',
        'reason': 'robustness test: stronger geometric/intensity augmentation without changing architecture',
        'env': {
            'UMUD_EPOCHS': '18',
            'UMUD_IMG_SIZE': '512',
            'UMUD_TTA_EXTRA_SIZE': '640',
            'UMUD_MODEL_ARCH': 'unet',
            'UMUD_MODEL_ENCODER': 'resnet34',
            'UMUD_MODEL_ENCODER_WEIGHTS': 'imagenet',
            'UMUD_LOSS_MODE': 'dice_bce',
            'UMUD_AUG_LEVEL': 'strong',
            'UMUD_BATCH_SIZE': '6',
            'UMUD_CLAHE': '0',
            'UMUD_FASC_POS_WEIGHT': '0',
            'UMUD_WEIGHTS_TAG': 'seg59_06',
        },
    },
]

COMMON_ENV = {
    # Keep the measurement/scale side explicit and comparable across runs.
    'UMUD_SCALE_ROUTER': '1',
    'UMUD_TTA': '1',
    'UMUD_FL_IDENTITY_BLEND': '0',
    'UMUD_FRAGMENT_FL': '1',
    'UMUD_USE_IDENTITY_FL': '1',
    'UMUD_FL_RECENTER': '1',
    'UMUD_TOP_BOUNDARY_MODE': 'line',
    'UMUD_MT_MODE': 'perp_center',
    'UMUD_FASC_MIN_AREA': '40',
    'UMUD_FASC_MIN_ANG': '6',
}

MAX_RUNS = 3
summaries = []
for run in RUNS[:MAX_RUNS]:
    run_id = run['run_id']
    print('\n' + '=' * 90)
    print('START', run_id)
    print(run['reason'])
    print('=' * 90)
    env = os.environ.copy()
    env.update(COMMON_ENV)
    env.update(run['env'])
    (WORK / f'run_config_{run_id}.json').write_text(json.dumps({'run_id': run_id, 'reason': run['reason'], 'env': {k: env[k] for k in sorted(set(COMMON_ENV) | set(run['env']))}}, indent=2), encoding='utf-8')

    t0 = time.time()
    status = 'ok'
    error = ''
    result = subprocess.run([sys.executable, 'segment_then_measure.py'], env=env, text=True)
    elapsed_min = (time.time() - t0) / 60.0
    if result.returncode != 0:
        status = 'failed'
        error = f'process returned {result.returncode}'
        print('FAILED', run_id, error)
    else:
        print('FINISHED', run_id, f'in {elapsed_min:.1f} min')

    produced = []
    for src_name, dst_name in [
        ('submission_segmentation.csv', f'submission_{run_id}.csv'),
        ('calibration_measurement_debug.csv', f'calibration_measurement_debug_{run_id}.csv'),
    ]:
        src = WORK / src_name
        if src.exists():
            dst = WORK / dst_name
            shutil.copy2(src, dst)
            produced.append(dst.name)

    stats = {}
    sub_path = WORK / f'submission_{run_id}.csv'
    dbg_path = WORK / f'calibration_measurement_debug_{run_id}.csv'
    if sub_path.exists():
        sub = pd.read_csv(sub_path)
        stats.update({
            'rows': int(len(sub)),
            'pa_mean': float(sub['pa_deg'].mean()),
            'pa_std': float(sub['pa_deg'].std()),
            'fl_mean': float(sub['fl_mm'].mean()),
            'fl_std': float(sub['fl_mm'].std()),
            'mt_mean': float(sub['mt_mm'].mean()),
            'mt_std': float(sub['mt_mm'].std()),
        })
        assert sub.shape == (309, 4), f'{run_id}: expected 309x4 submission, got {sub.shape}'
        assert sub['image_id'].is_unique, f'{run_id}: duplicate image_id rows'
        assert not sub[['pa_deg', 'fl_mm', 'mt_mm']].isna().any().any(), f'{run_id}: NaN predictions'
    if dbg_path.exists():
        dbg = pd.read_csv(dbg_path)
        if 'calibration_method' in dbg:
            stats['scaled_images'] = int((dbg['calibration_method'] != 'none').sum())
            stats['calibration_methods'] = dbg['calibration_method'].value_counts(dropna=False).head(12).to_dict()

    summaries.append({'run_id': run_id, 'status': status, 'elapsed_min': elapsed_min, 'error': error, 'produced': produced, **stats})
    pd.DataFrame(summaries).to_csv(WORK / 'seg59_sleep_matrix_summary.csv', index=False)
    print('Summary so far:')
    print(pd.DataFrame(summaries).to_string(index=False))

print('\nALL REQUESTED RUNS COMPLETE OR RECORDED.')
```

```python
# Bundle all submissions, calibration debug files, configs, summaries, and weights into one zip.
from pathlib import Path
from zipfile import ZipFile, ZIP_DEFLATED
from IPython.display import FileLink, display

WORK = Path('/kaggle/working')
zip_path = WORK / 'umud_seg59_sleep_matrix_outputs.zip'
patterns = [
    'submission_*.csv',
    'calibration_measurement_debug_*.csv',
    'seg59_sleep_matrix_summary.csv',
    'run_config_*.json',
    'seg_*.pt',
]
include_paths = []
for pattern in patterns:
    include_paths.extend(sorted(WORK.glob(pattern)))
include_paths = sorted({p.resolve(): p for p in include_paths if p.exists()}.values(), key=lambda p: p.name)

with ZipFile(zip_path, 'w', compression=ZIP_DEFLATED) as zf:
    for path in include_paths:
        zf.write(path, arcname=path.name)

print('Created:', zip_path)
print('Contents:')
for path in include_paths:
    print(' -', path.name, f'({path.stat().st_size:,} bytes)')

display(FileLink(str(zip_path)))
for path in include_paths:
    if path.name.startswith('submission_') or path.name == 'seg59_sleep_matrix_summary.csv':
        display(FileLink(str(path)))
```
